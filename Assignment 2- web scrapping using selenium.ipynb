# Question-1: Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

Connecting webdriver to Chrome Driver

```python
# Using chrome driver location 
driver1 = webdriver.Chrome(r"C:\Users\Prakash\Downloads\chromedriver_win 32 (1).zip\chromedriver.exe")
```

```python
# Getting the webpage ( webpage opened )
driver1.get('https://www.naukri.com/')
```

```python
# Entering the designation as 'Data Analyst' using class name

designation1 = driver1.find_element(By.CLASS_NAME, "suggestor-input")
designation1.send_keys('Data Analyst')
```

```python
# Entering the location as 'Bangalore' using Absolute XPATH(as finding a single element) 

loc1 = driver1.find_element(By.XPATH, "/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
loc1.send_keys('Bangalore')
```

```python
# Code for the clicking Search button using click()

search_button = driver1.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/div/div[6]')
search_button.click()
```

Scrapping the data for first 10 jobs:

```python
# Creating the empty lists to store the data 
Job_title1=[]
Job_loc1=[]
comp_name1=[]
exp_req1=[]
```

```python
# Code to scrape job_title for the first 10 jobs using Relative XPATH(as finding the multiple elements)

title_tags1 = driver1.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')

# loop for adding job_titles

for i in title_tags1[0:10]:
    Job_title1.append(i.text)   

# Code to scrape job_location for the first 10 jobs using Relative XPATH(as finding the multiple elements)

loc_tags1 = driver1.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')

# loop for adding job_Locations

for i in loc_tags1[0:10]:
    Job_loc1.append(i.text)   
    
# Code to scrape comapny_names for the first 10 jobs using Relative XPATH(as finding the multiple elements)

comp_tags1 = driver1.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')

# loop for adding company names

for i in comp_tags1[0:10]:
    comp_name1.append(i.text)      

# Code to scrape experience_required for the first 10 jobs using Relative XPATH(as finding the multiple elements)

exp_tags1 = driver1.find_elements(By.XPATH,'//span[@class="ellipsis fleft fs12 lh16 expwdth"]')

# loop for adding experience required

for i in exp_tags1[0:10]:
    exp_req1.append(i.text)       
```

```python
# Creating the DataFrame using pandas 
df1 = pd.DataFrame({"JOB_TITLE":Job_title1,"JOB_LOCATION":Job_loc1,"COMPANY_NAME":comp_name1,"EXPERIENCE_REQUIRED": exp_req1})
df1   
```

# Question-4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes: 1. Brand, 2. ProductDescription, 3. Price, 4. Offer

Connecting webdriver to Chrome Driver

```python
# Using chrome driver location 
driver4 = webdriver.Chrome(r"C:\Users\Prakash\Downloads\chromedriver_win 32 (1).zip\chromedriver.exe")
```

```python
# Getting the webpage ( webpage opened )
driver4.get("https://www.flipkart.com/")
```

```python
# Code for clicking cancel button 
cancel_button = driver4.find_element(By.XPATH,'/html/body/div[2]/div/div/button')
cancel_button.click()
```

```python
# Entering product as 'Sunglasses' using class name

product = driver4.find_element(By.CLASS_NAME, "_3704LK")
product.send_keys('Sunglasses')
```

```python
# Code for the clicking Search button using click()

search_button = driver4.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search_button.click()
```

SCRAPPING THE REQUIRED DATA FROM THE WEBPAGE

```python
# Creating lists to store data
brands=[]
product_desc=[]
price=[]
offer=[]
```

Code for scrapping the data of 100 sunglasses from 3 pages 



```python
# loop for redirecting the pages by clicking next button 

for page in range(0,3): # 3 pages contain the 100 sunglasses data 
   
 # code for getting the data of sunglasses of each page 
    
    brand_tags = driver4.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')   # brand_name
    price_tags = driver4.find_elements(By.XPATH,'//div[@class="_30jeq3"]')   # price 
    desc_tags = driver4.find_elements(By.CLASS_NAME,"IRpwTa")       # product_description
    offer_tags = driver4.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')   # offer 
    
 # single loop for adding the data to the lists

    for i in range(len(brand_tags)):
        
        brands.append(brand_tags[i].text)  # adding the brand name 
        
        product_desc.append(desc_tags[i].text)  # adding the product_description
        
        price.append(price_tags[i].text)    # adding the price
        
        offer.append(offer_tags[i].text)    # adding the offer
        
        # condition for scrapping data of only 100 sunglasses
        
        if(len(brands)>99):
            break

# code for switching to the next page by clicking next button

next_button = driver4.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]/span')
next_button.click()
time.sleep(3)
```

```python
# Creating the DataFrame using pandas 
df4=pd.DataFrame({"BRAND_NAME":brands,"PRODUCT_DESCRIPTION":product_desc,"PRICE":price,"OFFER":offer})
df4   
```

```python
# Code for clicking the 'LIST OF ALL PRIME MINISTERS OF INDIA' option 
driver9.find_element(By.XPATH,'/html/body/div[1]/div/div/div[2]/div/div[10]/div/div/ul/li[2]/a').click()
```

```python
# Creating lists to store the data 
name_p=[]
b_d=[]
term=[]
remarks=[]
```

```python
# Creating the DataFrame using pandas 
df9=pd.DataFrame({"NAME":name_p,"BORN-DEAD":b_d,"TERM OF OFFICE":term,"REMARKS":remarks})
df9   
```

```python
# Creating lists to store data 
name_c=[]
desc_c=[]
price_c=[]
```

```python
# Code for getting the data of 50 most expensive cars in the world 

#loop for getting Car names
    
name_tags=driver10.find_elements(By.XPATH,'//h3[@class="subheader"]') #  GETTING CAR NAMES
for i in name_tags[0:50]:
    name_c.append(i.text)
    
#loop for getting Car price

price_tags = driver10.find_elements(By.TAG_NAME,"strong")  # getting price 
for i in range(len(price_tags)):
    if(len(price_tags[i].text.split("\n")[0])>0):
        price_c.append(price_tags[i].text.split("\n")[0])
        
#loop for getting description of car 

desc_tags = driver10.find_elements(By.XPATH, '//p[not(@class) and text()!="NULL"]')
for i in desc_tags[3:53]:
        desc_c.append(i.text)
print(len(desc_c))
```
